#### Classification Decision Tree
- a supervised machine learning algorithm used to predict the class or category of a target variable based on a set of input features.
-  It is represented as a tree-like structure, where internal nodes represent decisions on a feature, branches represent the outcomes of those decisions, and leaf nodes represent the final class labels.

#### Type of problem
**Binary Classification**:
   - The target variable has two possible outcomes (e.g., "Yes" or "No").
   - Example: Predicting whether a customer will buy a product.
**Multi-Class Classification**:
   - The target variable has more than two categories.
   - Example: Classifying an animal as "Cat," "Dog," or "Rabbit."
**Key Features of Problems Addressed**
- The dataset contains both categorical and/or numerical input features.
- The relationship between input features and the target class may involve **non-linear decision boundaries**.
- Interpretability and decision-making transparency are important.

#### Real-world application

#### The differences between classification and regression
**type of problem addressed** (classification vs. regression) and the 
**approach used for prediction**. 
<br>
 **1. Type of Output**
- **Classification Decision Tree**:
  - Used for **classification** problems, where the output is a discrete category or class label (e.g., "Yes" or "No," "Cat" or "Dog").
  - Example: Predicting whether a loan application is "Approved" or "Denied."

- **Multiple Linear Regression**:
  - Used for **regression** problems, where the output is a continuous numerical value.
  - Example: Predicting the price of a house based on features like size, location, and age.
<br>

**2. Decision-Making Approach**
- **Classification Decision Tree**:
  - **Decision Process**:
    - The tree divides the input space into regions by making a series of binary splits based on feature values.
    - Splits are chosen to maximize the separation between classes using measures like **Gini Impurity** or **Entropy**.
  - **Prediction**:
    - A new data point follows the decision path down the tree.
    - At the leaf node, the class label is predicted based on the majority class of the training samples in that leaf.
    - If probabilities are needed, the proportions of samples in the classes at the leaf can be used.

- **Multiple Linear Regression**:
  - **Mathematical Model**:
    - Fits a linear equation to the data:
      \[
      y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \cdots + \beta_nx_n + \epsilon
      \]
      Where:
      - \( y \): Predicted continuous value.
      - \( x_1, x_2, \dots, x_n \): Input features.
      - \( \beta_0, \beta_1, \dots, \beta_n \): Coefficients of the model.
      - \( \epsilon \): Error term.
  - **Prediction**:
    - A new data point is input into the equation, and the predicted \( y \) value is calculated as a weighted sum of the feature values.

**3. Handling Non-Linear Relationships**
- **Classification Decision Tree**:
  - Can model **non-linear decision boundaries** by splitting the feature space recursively.
  - Example: If a boundary between classes is curved or irregular, the tree can approximate it with multiple splits.

- **Multiple Linear Regression**:
  - Assumes a **linear relationship** between the input features and the output variable.
  - Example: A house's price increases proportionally with its size, given all else remains constant.

**4. Interpretability**
- **Classification Decision Tree**:
  - Highly interpretable because it provides explicit decision rules (e.g., "If Age > 30 and Income < 50K, then Class = Yes").
  - Useful in domains like healthcare or finance, where interpretability is critical.

- **Multiple Linear Regression**:
  - Interpretable in terms of coefficients, which show the expected change in \( y \) for a one-unit change in \( x \), holding other variables constant.

**5. Example**
Suppose we have a dataset with **Age** and **Income** as features:

#### Predicting if a person buys a product (Classification Decision Tree):
- The tree might split like this:
  - If \( \text{Age} > 30 \):
    - If \( \text{Income} < 50K \): Predict "No."
    - Otherwise: Predict "Yes."
- Prediction: Follows the splits to arrive at a final class label.

#### Predicting how much a person will spend (Multiple Linear Regression):
- The model might output:
  \[
  \text{Spending} = 50 + 2 \cdot \text{Age} + 0.5 \cdot \text{Income}
  \]
- Prediction: For Age = 35 and Income = 40K, spending is:
  \[
  50 + 2(35) + 0.5(40) = 170
  \]

|--------------------------|------------------------------------------------------------|------------------------------------------|
| **Output**               | Discrete Class Label                                        | Continuous Numerical Value               |
| **Prediction Method**    | Recursive splitting; majority class in leaf node           | Weighted sum of input features           |

#### **1. Accuracy**
**Application Scenario**: **Quality control in manufacturing**  
- **Example**: Checking if products are "Defective" or "Non-defective."
- **Rationale**: When the cost of misclassification (false positives or false negatives) is low and the dataset is balanced, accuracy is a reliable measure. It provides a straightforward evaluation of overall model performance.
- High Accuracy means:
Most defective products are caught (high TP).
Most non-defective products are correctly passed (high TN).
- Low Accuracy could indicate:
Many defective products are missed (high FN).
Many non-defective products are unnecessarily flagged (high FP).
---

#### **2. Sensitivity (True Positive Rate)**
**Application Scenario**: **Disease screening in healthcare**  
- **Example**: Identifying patients with cancer from medical tests.
- **Rationale**: Sensitivity is critical when it is important to minimize false negatives (i.e., missing actual cases of the disease). Failing to identify a positive case could have severe consequences, such as delayed treatment.

---

#### **3. Specificity (True Negative Rate)**
**Application Scenario**: **Spam email detection**  
- **Example**: Classifying emails as "Spam" or "Not Spam."
- **Rationale**: Specificity is crucial when the focus is on minimizing false positives, such as misclassifying legitimate emails as spam. This ensures that the user does not miss important communications.

---

#### **4. Precision (Positive Predictive Value)**
**Application Scenario**: **Fraud detection in financial transactions**  
- **Example**: Detecting fraudulent credit card transactions.
- **Rationale**: Precision is essential when the cost of false positives is high (e.g., flagging genuine transactions as fraudulent). High precision ensures that only highly likely fraudulent cases are flagged, minimizing disruptions to legitimate users.

### Summary

Here’s a summary of our exchanges regarding the metrics and their applications, with examples to provide a clear and concise record:

---

### **Summary of Interactions**

1. **Definitions of TP, TN, FP, FN**:
   - We discussed how these terms apply to classification problems:
     - **True Positive (TP)**: Correctly predicting a defective product as defective.
     - **True Negative (TN)**: Correctly predicting a non-defective product as non-defective.
     - **False Positive (FP)**: Incorrectly predicting a non-defective product as defective.
     - **False Negative (FN)**: Incorrectly predicting a defective product as non-defective.

2. **Accuracy in Quality Control**:
   - **Accuracy** measures how well the model classifies both defective and non-defective products:
     \[
     \text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}
     \]
   - We concluded that accuracy evaluates the proportion of correct predictions (both TP and TN) relative to all predictions.

3. **Applications of Metrics**:
   - **Accuracy**: Useful when the dataset is balanced (similar number of defective and non-defective products) and the cost of errors (FP and FN) is similar.
   - **Limitations of Accuracy**: In cases of imbalanced datasets (e.g., if defective products are rare), accuracy may not be meaningful, as a model predicting "Non-defective" all the time could still have high accuracy but poor performance.

4. **When to Use Other Metrics**:
   - **Sensitivity**: Critical for minimizing **FN**, ensuring that defective products are rarely missed (e.g., in safety-critical scenarios).
   - **Specificity**: Focused on minimizing **FP**, useful when false alarms (flagging non-defective products as defective) are costly.
   - **Precision**: Important when we want high confidence in positive identifications, especially for minimizing unnecessary rework of good products.

---

### **Examples Provided**
- **Quality Control in Manufacturing**:
  - **TP**: A defective product is flagged as defective.
  - **TN**: A non-defective product is cleared as non-defective.
  - **FP**: A non-defective product is incorrectly flagged as defective.
  - **FN**: A defective product is incorrectly cleared as non-defective.

- **Model Accuracy in Context**:
  - High accuracy means the model is good at identifying both defective and non-defective products.
  - Low accuracy implies frequent misclassifications, either missing defective products or flagging non-defective ones unnecessarily.

**How Predictions Are Made in the Decision Tree:**
1. **Splits Based on Features**:
   - At each node, the tree decides a splitting condition using one of the predictor variables (`NumPages`, `Thick`, or `List Price`) and a threshold value.
   - For example:
     - "Is `List Price` ≤ $15?" → If yes, go left; if no, go right.

2. **Recursive Partitioning**:
   - The data is partitioned recursively at each node until a leaf node is reached or the maximum depth is met (in this case, `max_depth=4`).

3. **Leaf Nodes Contain Predictions**:
   - Leaf nodes represent the final classification decision.
   - They contain the class (e.g., "Paperback" or "Hardcover") and possibly additional information like the proportion of data points classified into each category at that node.

4. **Example Path**:
   - For a book with `NumPages=300`, `Thick=2.5`, and `List Price=$25`:
     - Start at the root node.
     - Check the first splitting condition (e.g., "Is `NumPages` ≤ 200?").
     - Continue down the tree until a leaf node is reached.
     - The class label at the leaf node is the prediction.

#### **1. What is a "positive" and what is a "negative"?**
- **Positive**: The class that the model is designed to identify or focus on. In the context of the `Hard_or_Paper` dataset:
  - A **positive** is a **Hardcover book** (denoted by `'Hard'`).
- **Negative**: The opposite class of the positive. In this case:

---

#### **2. How to read an sklearn confusion matrix?**
A confusion matrix summarizes the predictions of a classification model by comparing them to the true class labels. For binary classification, the structure is:

| **Actual/True Values** | **Predicted Positive** | **Predicted Negative** |
|-------------------------|------------------------|------------------------|
| **Actual Positive**     | **True Positive (TP)** | **False Negative (FN)**|
| **Actual Negative**     | **False Positive (FP)**| **True Negative (TN)** |

#### **Explanation of Cells**:
- **True Positive (TP)**: The model predicted the class as positive (Hardcover), and it was correct.
- **True Negative (TN)**: The model predicted the class as negative (Paperback), and it was correct.
- **False Positive (FP)**: The model predicted the class as positive (Hardcover), but it was wrong.
- **False Negative (FN)**: The model predicted the class as negative (Paperback), but it was wrong.

---

#### **3. What leads to TP, TN, FP, and FN?**
The outcomes depend on the interaction between true labels (`y_true`) and predicted labels (`y_pred`):

| True Label (`y_true`) | Predicted Label (`y_pred`) | Outcome              |
|-----------------------|---------------------------|----------------------|
| Positive (Hard)       | Positive (Hard)           | True Positive (TP)   |
| Negative (Paper)      | Negative (Paper)          | True Negative (TN)   |
| Negative (Paper)      | Positive (Hard)           | False Positive (FP)  |
| Positive (Hard)       | Negative (Paper)          | False Negative (FN)  |

---

#### **4. Does `y_true` or `y_pred` go first in `confusion_matrix`?**
In sklearn's `confusion_matrix` function, the order is:
```python
confusion_matrix(y_true, y_pred)
```
- **`y_true`**: True labels (ground truth from the dataset).
- **`y_pred`**: Predicted labels from the model.


````python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np

# Assuming clf and clf2 are already trained on the training data
# Define y_test as the true labels
y_test = pd.get_dummies(ab_reduced_noNaN_test["Hard_or_Paper"])['H']

# Define X_test for clf and clf2
X_test_clf = ab_reduced_noNaN_test[['List Price']]
X_test_clf2 = ab_reduced_noNaN_test[['NumPages', 'Thick', 'List Price']]

# Predictions for each model
y_pred_clf = clf.predict(X_test_clf)
y_pred_clf2 = clf2.predict(X_test_clf2)

# Confusion matrices for clf and clf2
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Function to calculate sensitivity, specificity, and accuracy
def calculate_metrics(cm):
    TP = cm[1, 1]  # True Positives
    TN = cm[0, 0]  # True Negatives
    FP = cm[0, 1]  # False Positives
    FN = cm[1, 0]  # False Negatives
    sensitivity = TP / (TP + FN) if (TP + FN) > 0 else 0  # Recall
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    return np.round(sensitivity, 3), np.round(specificity, 3), np.round(accuracy, 3)

# Calculate metrics for clf and clf2
metrics_clf = calculate_metrics(cm_clf)
metrics_clf2 = calculate_metrics(cm_clf2)

# Visualize confusion matrices
disp_clf = ConfusionMatrixDisplay(confusion_matrix=cm_clf, display_labels=["Paperback", "Hardcover"])
disp_clf2 = ConfusionMatrixDisplay(confusion_matrix=cm_clf2, display_labels=["Paperback", "Hardcover"])

disp_clf.plot()
disp_clf2.plot()

# Output metrics for clf and clf2
print("Metrics for clf (List Price only):", metrics_clf)
print("Metrics for clf2 (NumPages, Thick, List Price):", metrics_clf2)
````
