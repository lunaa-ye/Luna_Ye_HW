## Observations and Variables

An observation is an fact or figure we collect about a given variable. In the context of the dataset, the observation refers to each row which contains the information of the villager (such as its name, species, and personality).

A variable is any characteristic, number, or quantity that can be measured or counted to describe each observation. There are numerical variables (values with equal units and scales) and categorical variables (variable identifies category or group, not order or scale). In the context of the dataset, the variable refers to the columns. Names, species, birthdays, personality types, and other traits of the villagers are listed in the columns. These columns give you data points for each villager.

## The difference between an "attribute" and a "method"

Summary: In Python, attributes are variables that are a part of an object and hold data on its traits and properties. They can be used to depict information about the thing, such as facts or details. They can be accessed without (), and no action or computation is performed. Methods are functions that are connected to an object or class. They are employed to carry out particular actions or activities on the data kept in objects or classes. They can be called with (), and they can perform operations on the object.

## Definitions

Count: The number of non-missing (non-null) values in the column.

Mean: The average of the values. It is the sum of all values divided by the number of non-missing values.

Std: Standard deviation. A measure of how spread out the values are. It represents the average distance between each value and the mean.

Min: The smallest value in the column.

Max: The largest value in the column.

Median(Md): score corresponding to the point having 50% of the observations below it. It is also the second quartile (50%).

Queartiles divide an ordered (lowest to highest) dataset into quarters. 25%, 50%, and 75% are the first, second, and third quantiles referred to as  𝑄1, the median, and  𝑄3

## df.dropna() or del df['col']

Prefer df.dropna()

If the dataset contains missing values sporadically throughout multiple columns but only affects a small number of rows, we may prefer df.dropna() over df['col'] because it allows us to keep as many columns as possible and only remove rows that have missing data.

Example: 

````python
print(max(sb.isna().sum(axis=1)))
sb
````

In the sb dataset, for rows that have 2 missing values, we may choose to drop the rows rather than delete the columns.

Prefer del['col']

del['col'] may be chosen when a whole column has a large quantity of missing data. It would be preferable to eliminate the column in this situation in order to prevent adding bias or mistakes into the analysis.

Example:

````python
sb.isna().sum()
````

In the sb dataset, column 'deck' contains 688 missing values. We may choose to delete this column rather than drop rows.

Importance of Applying del df['col'] Before df.dropna(): Prior to using df.dropna(), it could be more effective to remove any columns that have a high number of missing data.
